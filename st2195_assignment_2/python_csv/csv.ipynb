## Introduction

In this Jupyter notebook we want to scrap the csv example on cars in the Wikipedia page and save the resulting output in the local folder 


### Load the library

library *requests* is to connect to the web server.

```python
from bs4 import BeautifulSoup
import requests
```

### Get the csv example

Load in the html

```python
wiki_html = requests.get("https://en.wikipedia.org/wiki/Comma-separated_values")
soup = BeautifulSoup(wiki_html.text, 'html.parser')
```

```python
wiki_html
```

```python
soup
```

Get the csv example under the header "Example"

```python
ths = soup.find(id='Example')  # locate where the id='Example' is at
csv_text = ths.findNext('pre').text  # get the first 'pre' right after id='Example'
csv_text
```

### Save the csv example into a csv file

Open a new file name 'car.csv', with 'w' (write) mode. <br>
write the *csv_text* extracted from the website, to the file.

```python
f = open('car.csv', 'w')
f.write(csv_text)
f.close()  # close the openned file to save it
```


Reload the csv data from the file to make sure the data was saved properly

```python
import pandas as pd
pd.read_csv('car.csv')
```
